```python
%sql
USE gold;
SELECT year, month, total_revenue, total_expenses, net_profit, ROUND(expense_ratio_pct, 2) AS expense_ratio_pct FROM financial_summary ORDER BY year DESC;



```

```python
%sql
USE gold;
WITH revenue AS (
  SELECT
    year,
    month,
    total_revenue,
    (year * 100 + month) AS ym
  FROM financial_summary
)
SELECT
  year,
  month,
  total_revenue,
  ROUND(
    (total_revenue - LAG(total_revenue) OVER (ORDER BY ym))
    / LAG(total_revenue) OVER (ORDER BY ym) * 100, 2
  ) AS mom_growth_pct
FROM revenue
ORDER BY year DESC, month DESC;
```

```python
%sql
USE gold;
SELECT department, month, budget_amount, actual_amount, variance, 
ROUND (try_divide(variance, budget_amount) *100, 2) AS variance_pct FROM budget_vs_actual WHERE variance > 0
ORDER BY variance DESC;
```

```python
%sql
USE gold;
SELECT payment_method, count(*) as transaction_count, ROUND(SUM(paid_amount), 2) AS total_collected
FROM sales_performance WHERE is_paid = 1
GROUP BY payment_method
ORDER BY total_collected DESC;
```
